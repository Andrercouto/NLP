### Roles da funo

Na funo *openai.ChatCompletion.create*, usada para interagir com o modelo GPT da OpenAI, o parmetro messages  uma lista que contm a troca de mensagens entre o usurio e o modelo. Dentro dessa lista, cada elemento  um dicionrio com duas chaves principais: **role** e **content**.

**O campo role dentro de cada item da lista messages indica o papel de quem est enviando a mensagem**. Ele pode ter os seguintes valores:

- "system": Este papel  usado para fornecer instrues iniciais ao modelo, que ajudam a configurar seu comportamento ou estilo de resposta. Geralmente,  usada para fornecer contexto sobre o que o modelo deve fazer (por exemplo, "Voc  um assistente til").

- "user": Este papel representa a mensagem enviada pelo usurio. Quando o usurio faz uma pergunta ou solicita algo ao modelo, a mensagem associada a esse papel descreve o pedido ou a interao.

- "assistant": Este papel representa a resposta do modelo para a mensagem do usurio. O contedo da mensagem associada a esse papel  a resposta gerada pelo modelo com base no que o usurio disse.

```python
# Testando respostas
response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "Voc  um jornalista de futebol."},
        {"role": "user", "content": "Quem foi o heroi da copa de 1962?"}
    ]
)
print(response.choices[0].message.content)
```

```python
# Testando respostas
response2 = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "Voc  um especialista em NLP."},
        {"role": "user", "content": "Explique-me o funcionamento de uma estrutura multihead attention."}
    ]
)
print(response2.choices[0].message.content)
```
