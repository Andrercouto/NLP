# **BERT - Bidirectional Encoder Representation from Transformer**

BERT é um modelo Transformer que utiliza apenas a parte do **Encoder**, focado em gerar representações linguísticas ricas e contextualizadas. Por ser **bidirecional**, ele processa o texto em ambas as direções, permitindo capturar relações contextuais entre as palavras.

---

## **Características principais**
- **Representação contextualizada**: As embeddings das palavras são ajustadas conforme o contexto, utilizando o mecanismo de **Multi-head Attention**.
- **Pré-treinado**: O modelo é pré-treinado em grandes corpora (**Toronto BookCorpus** e **Wikipedia**), evitando o alto custo de treinar do zero.
- **Treinamento não supervisionado**: Não exige rótulos, pois utiliza tarefas auto-supervisionadas durante o treinamento inicial.
- **Flexível**: Pode ser adaptado (**fine-tuned**) para tarefas específicas, mesmo em outros idiomas.



## **Tarefas de pré-treinamento**

#### Masked Language Modeling (MLM)
- **15%** dos tokens são mascarados durante o treinamento.
- O modelo tenta prever os tokens mascarados com base no contexto (palavras vizinhas).
- Ajusta os pesos por meio de previsões probabilísticas para as palavras ausentes.

#### Next Sentence Prediction (NSP)
- O modelo trabalha com pares de sentenças e classifica se a segunda é uma continuação lógica da primeira.
- Trata-se de uma tarefa binária: **isNext** (continuação) ou **notNext** (não é continuação).


## **Funcionamento técnico**
Antes de processar o texto, o BERT converte a entrada em três camadas de embeddings:

#### **1. Token Embedding**
- Representa as palavras ou subpalavras (tokens).
- Utiliza tokens especiais:
  - `[CLS]`: adicionado no início da entrada, utilizado para classificações globais.
  - `[SEP]`: marca o final de cada sentença.
- Palavras desconhecidas são divididas em subpalavras sinalizadas com `##`.

#### 2. Segment Embedding
Identifica sentenças diferentes dentro da entrada de forma que cada segmento receba um marcador:
- Segment A: primeira sentença.
- Segment B: segunda sentença.

#### 3. Positional Embedding

- Indica a posição de cada token na sequência.
- Essencial para que o modelo reconheça a ordem dos tokens, já que o processamento é paralelo.
