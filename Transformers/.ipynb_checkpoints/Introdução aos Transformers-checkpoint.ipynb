# Transformers

Os Transformers representam o estado da arte em tecnologia para processamento de linguagem natural (NLP). Introduzidos pelo paper "Attention is All You Need" em 2017, essa arquitetura revolucionou o campo ao usar mecanismos de ateno para capturar relaes contextuais profundas entre palavras, frases e at mesmo documentos inteiros.

Uma de suas principais vantagens  a possibilidade de utilizar modelos pr-treinados, como BERT e GPT, que podem ser ajustados (fine-tuning) para tarefas especficas, como anlise de sentimentos, traduo automtica ou gerao de texto. Essa flexibilidade tornou os Transformers indispensveis em diversas aplicaes, desde chatbots at sistemas de busca inteligentes.

Os Transformers utilizam uma arquitetura baseada em dois componentes principais: Encoder e Decoder (ENDEC). O Encoder processa a entrada, convertendo-a em uma representao interna que o modelo pode entender, capturando relaes contextuais entre as palavras. O Decoder, por sua vez, utiliza essa representao para gerar a sada, como a traduo de um texto ou a continuao de uma frase. Essa estrutura modular  altamente flexvel e pode ser adaptada para diferentes tarefas, seja utilizando apenas o Encoder (como no BERT), apenas o Decoder (como no GPT) ou ambos em conjunto (como no Transformer original).

## Modelos de Ateno

Os mecanismos de ateno so componentes fundamentais dos modelos Transformers. Eles permitem que o modelo foque seletivamente em partes especficas de uma sequncia de entrada (como palavras em uma frase) ao process-la.

A ideia principal  atribuir pesos diferentes a cada elemento da sequncia, dependendo de sua relevncia para a tarefa em questo. Isso ajuda o modelo a captar melhor as relaes contextuais entre palavras, independentemente da distncia entre elas na frase.

## Encoders

### Embedding e Positional Encoder



### Embedding

A ideia  transformar representaes simblicas (como palavras) em vetores que possam ser usados pelo modelo. Esses vetores contm informaes semnticas e sintticas sobre as palavras, permitindo que o modelo entenda suas relaes. 

No embedding, a palavra  representada como um vetor baseado em seu significado geral, independente do contexto especfico da frase em que aparece.  como um "significado bsico" da palavra no idioma, antes de levar em conta como ela interage com outras palavras na frase.

### Positional Encoder

 uma matriz que vai informar para o processo do transformer a posio das palavras (uma vez que processamento de linguagem natural necessita de ordem). A maioria das tcnicas dependem de recorrncia, mas o **positional encoder** submete todo o texto **em paralelo**, tornando-o mais performtico e melhorando o aprendizado (j que diversos mecanismos de attention sero treinados simultaneamente). Essa tcnica utiliza funes matemticas, como senos e cossenos, para representar as posies, o que facilita o aprendizado dos mecanismos de ateno treinados simultaneamente.


### Embedding + Positional Encoder

No Transformer, o embedding e o positional encoding so representados como matrizes do mesmo tamanho, e esses dois componentes so somados para criar a entrada final que ser processada pelo modelo. Vamos detalhar:

- Matriz Embedding :
Cada palavra (ou token)  transformada em um vetor de dimenso fixa, digamos  (por exemplo, 512 ou 768 dimenses). Se o texto tem N palavras, o embedding gerar uma matriz de dimenso N x .

 - Positional Encoding Matrix: O positional encoding tambm gera uma matriz de dimenso N x , onde N  o nmero de palavras na sequncia e  o mesmo tamanho usado no embedding. Isso garante que ambas as matrizes possam ser somadas diretamente.

A soma  feita posio por posio (element-wise addition) na matriz. Assim, para cada palavra/token, o vetor embedding da palavra  combinado com o vetor do positional encoding que indica sua posio na sequncia.



Um modelo Transformer pode ter N encoders, ou seja, ele pode ser composto por mltiplas camadas de Encoder no seu lado esquerdo (lado de entrada). Cada Encoder processa a entrada em sequncia, e o nmero de encoders  um hiperparmetro que pode ser ajustado durante o design do modelo.



### Self Attention

A primeira etapa dos Encoders  o **Multi-head Attention**, porm, para isso  necessrio entender o **Self Attention**. O Self Attention  o mecanismo principal de um transformer. Esse mecanismo tem por funcao descobrir a relacao das palavras, por exemplo, na frase:

The Animal didnt cross the street because it was to wide.

Ns conseguimos saber que o it nesse caso est relacionado  rua, mas o computador no sabe se o it corresponde  street ou animal, mas o transformer aclcula a representacao e a relacao das palavras, podendo assim compreender que o it corresponde a rua.

Ele descobre essa relao atravs de 5 etapas:

1. **Matrizes Q, K e V**: Para cada palavra na frase, so gerados trs vetores: Query (Q), Key (K) e Value (V). Esses vetores so derivados dos embeddings das palavras atravs de multiplicaes com matrizes de pesos. O Query representa a palavra que est tentando se conectar com outras palavras, o Key ajuda a medir a similaridade entre as palavras e o Value contm a informao que ser passada para o prximo passo.
    
    As matrizes de pesos para Q, K e V comeam com valores aleatrios no incio do treinamento e so ajustadas gradualmente  medida que o modelo aprende. Durante o treinamento, o modelo otimiza essas matrizes para que as representaes de Q, K e V ajudem a identificar as relaes mais relevantes entre as palavras no contexto.

2. **Score: Produto escalar entre Query e Key**: O prximo passo  calcular a ateno. Para isso, o modelo calcula a similaridade entre o Query de uma palavra e o Key de todas as outras palavras. Isso  feito por meio de um produto escalar entre o Query e os Keys das outras palavras, seguido por uma normalizao (geralmente com a funo softmax). O resultado dessa operao  uma medida de quanta ateno uma palavra deve prestar a cada outra palavra na frase.

3. **Diviso de QK pela raiz quadrada das dimenes de Key**: Aps calcular o produto escalar entre o Query (Q) e o Key (K), a prxima etapa  dividir o valor obtido pela raiz quadrada do nmero de dimenses de Key (ou o tamanho do vetor de Key). Isso  feito para evitar que os valores do produto escalar cresam muito  medida que as dimenses aumentam, o que poderia afetar a estabilidade do treinamento. Essa diviso ajuda a manter os valores de ateno em uma faixa controlada e contribui para uma melhor normalizao, ajudando o modelo a aprender de maneira mais eficiente.
 
4. **Produo da Score Matrix**: Aps calcular a similaridade entre os vetores Query e Key, e normalizar os valores atravs da diviso pela raiz quadrada da dimenso de Key, o prximo passo  criar a Score Matrix. A Score Matrix contm os resultados da ateno calculada entre todas as palavras na sequncia, com cada valor indicando a fora da ateno entre duas palavras especficas.

    Essa matriz  o produto das operaes de ateno de todas as palavras na entrada, representando quanto cada palavra deve se concentrar nas demais palavras. A Score Matrix  essencial para medir a relao de dependncia entre as palavras dentro da sequncia. Ela  a base para a prxima etapa do clculo de ateno, onde ser aplicada a funo softmax para transformar esses valores em probabilidades.

5. **Calculo da Matriz Z**: A partir da Score Matrix gerada na etapa anterior, a prxima fase  calcular a Matriz Z. A Matriz Z  obtida multiplicando a Score Matrix pela Value Matrix (V). A Matriz Z contm as representaes ponderadas das palavras, com base nas atenes calculadas na Score Matrix. Ou seja, cada palavra da sequncia agora tem uma representao que leva em considerao as palavras mais relevantes para ela, ponderadas pelas suas atenes. A ideia  que as palavras mais importantes (aqueles que tm maior ateno) influenciem mais a representao final, enquanto as palavras com menos ateno tm um impacto menor.


$$
\text{Attention Weights} = \text{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)
$$

Nesse ponto, a matriz Z est[a pronta para ser enviada atravs da prxima camada Feedfoward Network (porm, antes ela ser normalizada).

### Multi-head Attention
No Multi-head Attention, o processo de Self Attention  realizado vrias vezes, com cada cabea produzindo um resultado que  somado. As matrizes resultantes so ento concatenadas e multiplicadas por uma matriz ponderada, gerando uma nica matriz de sada. Esse processo permite ao modelo capturar diferentes aspectos da ateno, melhorando a preciso e a expressividade do sistema.

### Add & Norm
Aps a concatenao e multiplicao das cabeas no Multi-head Attention, a sada  somada com a entrada original do bloco, formando um resduo. Em seguida, a soma  normalizada com a tcnica de Layer Normalization. Esse processo de Add & Norm tem o objetivo de estabilizar o treinamento e melhorar a convergncia do modelo.

### Feedforward Network
A Feedforward Network  composta por duas camadas densas intercaladas com funes de ativao ReLU. Essas camadas so responsveis por aplicar transformaes no-lineares nas representaes das palavras, permitindo ao modelo capturar relaes mais complexas.

A sada do primeiro encoder vai servir como entrada do encoder dois, e assim por diante.

### Simulao do processo de Self Attention

```python
import numpy as np

# Definindo os vetores de embeddings para 3 palavras
tokens = ['The', 'Animal', 'Street']

# Passo 1 - Matrizes Q, K e V
# Vetores de embedding (sempre inicializados aleatoriamente)
Q = np.random.randn(3, 4)  # Query matrix com 3 palavras, 4 dimenses
K = np.random.randn(3, 4)  # Key matrix
V = np.random.randn(3, 4)  # Value matrix

# Passo 2 - Produto escalar entre Q e K (transposta)
raw_attention_scores = np.dot(Q, K.T)

# Passo 3 - Diviso pela raiz quadrada do tamanho da dimenso de Key (sqrt(d_k))
d_k = K.shape[1]  # Tamanho da dimenso de K

# Passo 4 - Produo da Score Matrix (com normalizao)
score_matrix = raw_attention_scores / np.sqrt(d_k)

# Passo 5 - Clculo da Matriz Z:
# Funo softmax
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=1, keepdims=True)

# Aplicando softmax para normalizar as atenes
attention_weights = softmax(score_matrix)

# Obtendo a matriz Z (ponderao das palavras pela ateno)
Z = np.dot(attention_weights, V)

# Resultados
print("Query (Q):\n", Q)
print("\nKey (K):\n", K)
print("\nValue (V):\n", V)
print("\nRaw Attention Scores:\n", raw_attention_scores)  # Matriz de pontuaes brutas
print("\nScore Matrix (normalizada):\n", score_matrix)  # Matriz de pontuaes normalizadas
print("\nAttention Weights (Softmax):\n", attention_weights)  # Pesos de ateno aps softmax
print("\nFinal Output (Z):\n", Z)  # Resultado final ponderado pelas atenes
```

### Simulao do processo de Multi-head Attention

```python
# Funo softmax
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=1, keepdims=True)

# Definindo as matrizes Q, K, V e suas dimenses
tokens = ['The', 'Animal', 'Street']
n_heads = 2  # Nmero de cabeas de ateno (simulando 2 cabeas)

# Vetores de embedding (sempre inicializados aleatoriamente)
Q = np.random.randn(3, 4)  # Query matrix
K = np.random.randn(3, 4)  # Key matrix
V = np.random.randn(3, 4)  # Value matrix

# Projees de Q, K, V para mltiplas cabeas
Q_heads = np.random.randn(n_heads, 3, 4 // n_heads)  # Dividindo em cabeas
K_heads = np.random.randn(n_heads, 3, 4 // n_heads)  # Dividindo em cabeas
V_heads = np.random.randn(n_heads, 3, 4 // n_heads)  # Dividindo em cabeas

# Funo para calcular a ateno para cada cabea
def multi_head_attention(Q, K, V):
    attention_outputs = []
    
    for i in range(n_heads):
        raw_attention_scores = np.dot(Q_heads[i], K_heads[i].T)
        
        d_k = K_heads[i].shape[1]
        score_matrix = raw_attention_scores / np.sqrt(d_k)
        
        attention_weights = softmax(score_matrix)
        
        Z = np.dot(attention_weights, V_heads[i])
        
        attention_outputs.append(Z)
    
    output = np.concatenate(attention_outputs, axis=-1)
    
    final_output = np.dot(output, np.random.randn(4, 4))
    
    return final_output

# Obtendo a sada do multi-head attention
multi_head_output = multi_head_attention(Q, K, V)

# Resultados
print("Final Output (Multi-Head Attention):\n", multi_head_output)
```

## Decoders

O Decoder utiliza camadas empilhadas com embeddings e *Positional Encoding* para processar sequncias de sada. No entanto, sua principal diferena em relao ao Encoder  o uso do **Masked Multi-head Attention**, que impede que o modelo considere palavras futuras durante a gerao da sequncia.

Cada camada do Decoder recebe duas entradas:

- A sada da camada anterior do Decoder (ou um token inicial, <sos>, na primeira iterao).
- A sada final do Encoder, que contm o contexto da sequncia de entrada.



Durante o processo, o Decoder aplica o **Masked Multi-head Attention**, que mascara as palavras subsequentes da sequncia em gerao. Isso permite que o modelo gere as palavras uma de cada vez, garantindo causalidade. A entrada inicial  composta pelo token <sos>, enquanto o token <eos> marca o fim da sequncia.

No final de cada camada, a sada do Decoder passa por uma camada linear, que projeta o vetor em um espao de dimenso igual ao vocabulrio. Por fim, o softmax converte essa projeo em uma distribuio de probabilidades, permitindo que o modelo selecione a palavra com a maior probabilidade para continuar a sequncia.
